### Exercise 1:
Write a program which will find all such numbers which are symetric (such as 11, 121, 1221,...) between 10 and 2000 (both included). The numbers obtained should be printed in a commaseparated sequence on a single line.

```python
def symnum(a, b):
    if (type(a)!=int) or (type(b)!=int):
        print('Please enter two integers.')
    else:
        i=a
        while(i<(b+1)):
            m=i
            n=i
            rev=0
            while(n>0):
                ld=n%10
                rev=rev*10+ld
                n=n//10
            if(m==rev):
                print(i, end=', ')
            i=i+1
```

```python
symnum(10, 2000)
```

### Exercise 2:
Write a program which will find all such numbers which include at least 3 duplicated digits (such as 111, 1222,...) between 100 and 2000 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
def duplicated_digit(num):
    num_list=list(str(num))
    unique_value= np.unique(np.array(num_list))
    count_num =[]
    for i in unique_value:
        count_num.append(num_list.count(i))
    max_num=max(count_num)
    return_dict= {'distinct value':unique_value,'count':count_num, 'max': max_num}
    return return_dict
```

```python
for i in np.linspace(100, 2000, 1901):
    if duplicated_digit(int(i))['max']>=3:
        print(int(i),end=', ')
```

### Exercise 6:
Write a program which will find all primes between 1000 and 3000 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line. 

```python
def primes(a, b):
    if (type(a)!=int) or (type(b)!=int) or (b<=a):
        print('Please enter two integers in ascending order.')
    else:
        print('The prime between ', a, ' and ', b, ': ')
        for i in range(a, b+1):
            prime=True
            for j in range(2, i):
                if(i%j==0):
                    prime=False
            if prime:
                print(i, end=', ')
```

```python
primes(1000, 3000)
```

### Exercise 7:
Write a program which will receive an integer larger than 0 and respond all perfect numbers smaller than that integer. A perfect number is sum of its divisors except itself, such as 6 = 1+2+3. The numbers obtained should be printed in a comma-separated sequence on a single line. 

```python
def perfnum(x):
    if(type(x)!=int) or (x<=0):
        print('Please enter an iteger larger than 0.')
    else:
        n=1
        while(n<x):
            sum1=0
            for i in range(1, n):
                if(n%i==0):
                    sum1=sum1+i
            if(sum1==n):
                print(n, end=', ')
            n=n+1
```

```python
perfnum(2508)
```

### Exercise 8:
With a given integral number n, write a program to generate a dictionary that contains (i, i^3) such that is an integral number between 1 and n (both included). and then the program should print the dictionary. Suppose the following input is supplied to the program: 6. Then, the output should be: {1: 1, 2: 8, 3: 27, 4: 64, 5: 125, 6: 216}

```python
def dupdict(x):
    if(type(x)!=int) or (x<=1):
        print('Please enter an iteger larger than 1.')
    else:
        dict1={}
        for i in range(1, x+1):
            dict1[i]=int(i)**3
        print(dict1)
```

Write a program which accepts a sequence of comma separated 4 digit binary numbers as its input and then check whether they are divisible by 3 or not. The numbers that are divisible by 3 are to be printed in a comma separated sequence. Example: 0100,0011,1010,1101 Then the output should be: 1101 

```python
str1=str(input('Enter 4 comma-seperated binary numbers: '))
lst1=str1.split(',')
lst=[]
for i in lst1:
    lst.append(int(i))
print(lst)
j=0
print('The binary numbers divisible by 3: ')
while(j<len(lst)):
    if(lst[j]%3==0):
        print(lst[j], end=', ')
    j=j+1
```

### Exercise 10:
Write a program which will receive an integer larger than 0 and respond if that integer is sum of three distinct squared numbers, such as 38 = 22 + 32 + 52.

```python
def sum_of_square(number):
    square=[]
    for i in range(1,(int(number**(1/2)))+1):
        square.append(i**2)
    combi_list=list(C(square,3))
    sum_n=[x+y+z for (x,y,z) in combi_list]
    n=0
    for i in range(len(sum_n)):
        if sum_n[i]==number:
            n=n+1
    return n
```

```python
number=float(input('Enter a positive integer: '))
while number!= int(number) or number<= 0:
    number=float(input("Please enter a positive integer: "))
number=int(number)

if sum_of_square(number)==1:
    print("The input number is a sum of 3 distinct square numbers.")
else:
    print("The input number is not a sum of 3 distinct square numbers")
```
